### Deliverable 1: Preprocessing the Data for PCA

```python
# Load the crypto_data.csv dataset.
# YOUR CODE HERE
file_path = "./crypto_data.csv"
crypto_df = pd.read_csv(file_path)
crypto_df.head(10)
```

```python
# Keep all the cryptocurrencies that are being traded.
# YOUR CODE HERE
crypto_cleaning_df = crypto_df.loc[crypto_df.IsTrading == True,:]
crypto_cleaning_df.head(20)
```

```python
# Keep all the cryptocurrencies that have a working algorithm.
# YOUR CODE HERE
crypto_cleaning_df.Algorithm.unique()
```

```python
# Remove the "IsTrading" column. 
# YOUR CODE HERE
crypto_cleaning_df = crypto_cleaning_df.drop("IsTrading", axis = 1)
```

```python
# Remove rows that have at least 1 null value.
# YOUR CODE HERE
crypto_cleaning_df = crypto_cleaning_df.dropna()
crypto_cleaning_df
```

```python
# Keep the rows where coins are mined.
# YOUR CODE HERE
crypto_cleaning_df = crypto_cleaning_df.loc[crypto_df.TotalCoinsMined > 0]
```

```python
# Create a new DataFrame that holds only the cryptocurrencies names.
# YOUR CODE HERE
coinName_df = crypto_cleaning_df[["Unnamed: 0","CoinName"]]
coinName_df = coinName_df.set_index("Unnamed: 0")
coinName_df
```

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm.
# YOUR CODE HERE
crypto_cleaning_df.drop(["CoinName", "Unnamed: 0"], axis = 1, inplace = True)

crypto_cleaning_df.head()
```

```python
# Use get_dummies() to create variables for text features.
# YOUR CODE HERE 
X = crypto_cleaning_df.drop(["TotalCoinsMined","TotalCoinSupply"], axis = 1)
X = pd.get_dummies(X)
X.head()
```

```python
# Standardize the data with StandardScaler().
# YOUR CODE HERE
scaler = StandardScaler()
crypto_scaled = scaler.fit_transform(crypto_cleaned_df)
```

### Deliverable 2: Reducing Data Dimensions Using PCA

```python
# Using PCA to reduce dimension to three principal components.
# YOUR CODE HERE
pca = PCA(n_components=3)
pca.fit(crypto_scaled)
print(pca.explained_variance_ratio_)
crypto_pca = pca.fit_transform(crypto_scaled)
```

```python
# Create a DataFrame with the three principal components.
# YOUR CODE HERE
crypto_pca_df = pd.DataFrame(
    data = crypto_pca, columns=["principal component 1","principal component 2","principal component 3"]
)
crypto_pca_df.head()
```
